```python
import moviepy.editor as mp
import os
import wave
import json
from vosk import Model, KaldiRecognizer

def extract_audio_from_video(video_path, audio_path):
    # Load the video file
    video = mp.VideoFileClip(video_path)
    # Extract audio
    video.audio.write_audiofile(audio_path)

def audio_to_text(audio_path, model_path):
    # Load the Vosk model
    model = Model(model_path)
    # Recognizer
    recognizer = KaldiRecognizer(model, 16000)

    # Open the audio file
    wf = wave.open(audio_path, "rb")
    if wf.getnchannels() != 1:
        raise ValueError("Audio file must be mono")
    if wf.getsampwidth() != 2:
        raise ValueError("Audio file must be 16-bit")
    if wf.getframerate() != 16000:
        raise ValueError("Audio file must have a sample rate of 16000")

    # Process the audio file and perform recognition
    text = ""
    while True:
        data = wf.readframes(4000)
        if len(data) == 0:
            break
        if recognizer.AcceptWaveform(data):
            result = recognizer.Result()
            result_dict = json.loads(result)
            text += result_dict.get("text", "") + " "

    # Final result
    final_result = recognizer.FinalResult()
    final_result_dict = json.loads(final_result)
    text += final_result_dict.get("text", "")

    return text

video_path = "testvideo.mp4"
audio_path = "extracted_audio.wav"
model_path = "path_to_vosk_model"

# Extract audio from video
extract_audio_from_video(video_path, audio_path)

# Convert audio to text
text = audio_to_text(audio_path, model_path)
print("Extracted Text: ")
print(text)
```

```python
import moviepy.editor as mp
import os
from pocketsphinx import AudioFile, get_model_path, Pocketsphinx

def extract_audio_from_video(video_path, audio_path):
    try:
        video = mp.VideoFileClip(video_path)
        video.audio.write_audiofile(audio_path)
    except Exception as e:
        print(f"Error extracting audio: {str(e)}")
        raise

def audio_to_text(audio_path):
    try:
        # Get the default model path
        model_path = get_model_path()
        
        # Basic configuration
        config = {
            'verbose': True,  # Set to True for debugging
            'audio_file': audio_path,
        }
        
        # Add model paths if they exist
        hmm_path = os.path.join(model_path, 'en-us')
        if os.path.exists(hmm_path):
            config['hmm'] = hmm_path
        
        # Initialize Pocketsphinx
        ps = Pocketsphinx(**config)
        
        # Perform speech recognition
        text = ""
        with open(audio_path, 'rb') as audio_file:
            ps.decode_raw(audio_file)
            text = ps.hypothesis()
        
        return text
    except Exception as e:
        print(f"Error in speech recognition: {str(e)}")
        raise

video_path = "testvideo.mp4"
audio_path = "extracted_audio.wav"

try:
    # Extract audio from video
    extract_audio_from_video(video_path, audio_path)
    
    # Convert audio to text
    text = audio_to_text(audio_path)
    
    print("Extracted Text: ")
    print(text)

except Exception as e:
    print(f"An error occurred: {str(e)}")
    print("Please make sure you have the latest version of pocketsphinx installed:")
    print("pip install --upgrade pocketsphinx")
```

```python
import moviepy.editor as mp
import os
from pocketsphinx import Decoder, get_model_path
import speech_recognition as sr

def extract_audio_from_video(video_path, audio_path):
    try:
        video = mp.VideoFileClip(video_path)
        video.audio.write_audiofile(audio_path)
    except Exception as e:
        print(f"Error extracting audio: {str(e)}")
        raise

def audio_to_text(audio_path):
    try:
        # Get the default model path
        model_path = get_model_path()
        
        # Configuration for the decoder
        config = Decoder.default_config()
        config.set_string('-hmm', os.path.join(model_path, 'en-us'))
        config.set_string('-lm', os.path.join(model_path, 'en-us.lm.bin'))
        config.set_string('-dict', os.path.join(model_path, 'cmudict-en-us.dict'))
        
        # Initialize the decoder
        decoder = Decoder(config)
        
        # Decode the audio file
        with open(audio_path, 'rb') as audio_file:
            decoder.start_utt()
            while True:
                buf = audio_file.read(1024)
                if buf:
                    decoder.process_raw(buf, False, False)
                else:
                    break
            decoder.end_utt()
        
        # Get the hypothesis
        hypothesis = decoder.hyp()
        if hypothesis:
            return hypothesis.hypstr
        else:
            return "No speech detected"

    except Exception as e:
        print(f"Error in speech recognition: {str(e)}")
        raise

def audio_to_text_backup(audio_path):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio = recognizer.record(source)
    try:
        return recognizer.recognize_google(audio)
    except sr.UnknownValueError:
        return "Google Speech Recognition could not understand audio"
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"

video_path = "testvideo.mp4"
audio_path = "extracted_audio.wav"

try:
    # Extract audio from video
    extract_audio_from_video(video_path, audio_path)
    
    # Try PocketSphinx first
    print("Attempting speech recognition with PocketSphinx...")
    text = audio_to_text(audio_path)
    
    # If PocketSphinx fails, use Google Speech Recognition as a backup
    if not text or text == "No speech detected":
        print("PocketSphinx failed. Attempting with Google Speech Recognition...")
        text = audio_to_text_backup(audio_path)
    
    print("Extracted Text: ")
    print(text)

except Exception as e:
    print(f"An error occurred: {str(e)}")
    print("Please make sure you have the latest versions of pocketsphinx and SpeechRecognition installed:")
    print("pip install --upgrade pocketsphinx SpeechRecognition")
```

```python
import moviepy.editor as mp
import json
import os
from vosk import Model, KaldiRecognizer
import wave

def extract_audio_from_video(video_path, audio_path):
    try:
        print("start")
        video = mp.VideoFileClip(video_path)
        video.audio.write_audiofile(audio_path, codec='pcm_s16le', ffmpeg_params=["-ac", "1"])
        print("End")
    except Exception as e:
        print(f"Error extracting audio: {str(e)}")
        raise

def audio_to_text(audio_path, model_path):
    try:
        print("Loading Vosk model...")
        model = Model(model_path)
        
        print("Performing speech recognition...")
        rec = KaldiRecognizer(model, 16000)

        with wave.open(audio_path, 'rb') as wf:
            while True:
                data = wf.readframes(400)
                if len(data) == 0:
                    break
                if rec.AcceptWaveform(data):
                    pass

        result = json.loads(rec.FinalResult())
        return result['text']

    except Exception as e:
        print(f"Error in speech recognition: {str(e)}")
        raise

video_path = "AI & ML in Finance - Lecture 7 - Data Generation.mp4"
audio_path = "extracted_audio.wav"
model_path = "vosk-model-small-en-us-0.15"  # You need to download this

try:
    # Extract audio from video
    extract_audio_from_video(video_path, audio_path)
    
    # Convert audio to text
    text = audio_to_text(audio_path, model_path)
    
    print("Extracted Text: ")
    print(text)

except Exception as e:
    print(f"An error occurred: {str(e)}")
    print("Please make sure you have the required libraries and model:")
    print("pip install moviepy vosk")
    print("Download a Vosk model from: https://alphacephei.com/vosk/models")
finally:
    # Ensure the audio file is removed even if an error occurs
    if os.path.exists(audio_path):
        os.remove(audio_path)
```

```python
import os
import subprocess
from pocketsphinx import Pocketsphinx, get_model_path
from pydub import AudioSegment
from pydub.silence import split_on_silence
import concurrent.futures

def extract_audio_from_video(video_path, audio_path):
    try:
        print("Extracting audio...")
        cmd = [
            "ffmpeg", "-i", video_path,
            "-acodec", "pcm_s16le",
            "-ac", "1",
            "-ar", "16000",
            "-vn", audio_path
        ]
        subprocess.run(cmd, check=True, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        print("Audio extraction complete.")
    except subprocess.CalledProcessError as e:
        print(f"Error extracting audio: {str(e)}")
        raise

def enhance_audio(input_path, output_path):
    try:
        print("Enhancing audio...")
        cmd = [
            "ffmpeg", "-i", input_path,
            "-af", "highpass=f=200,lowpass=f=3000,afftdn=nf=-25",
            "-acodec", "pcm_s16le",
            "-ac", "1",
            "-ar", "16000",
            output_path
        ]
        subprocess.run(cmd, check=True, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        print("Audio enhancement complete.")
    except subprocess.CalledProcessError as e:
        print(f"Error enhancing audio: {str(e)}")
        raise

def transcribe_audio_chunk(audio_chunk, ps):
    raw_data = audio_chunk.raw_data
    ps.start_utt()
    ps.process_raw(raw_data, False, True)
    ps.end_utt()
    return ps.hypothesis()

def audio_to_text(audio_path):
    try:
        print("Performing speech recognition...")
        
        # Configure PocketSphinx
        model_path = get_model_path()
        ps_config = {
            'hmm': os.path.join(model_path, 'en-us'),
            'lm': os.path.join(model_path, 'en-us.lm.bin'),
            'dict': os.path.join(model_path, 'cmudict-en-us.dict')
        }
        ps = Pocketsphinx(**ps_config)
        
        # Load audio file
        audio = AudioSegment.from_wav(audio_path)
        
        # Split audio on silence
        chunks = split_on_silence(audio, min_silence_len=500, silence_thresh=audio.dBFS-14)
        
        # Transcribe chunks in parallel
        with concurrent.futures.ThreadPoolExecutor() as executor:
            futures = [executor.submit(transcribe_audio_chunk, chunk, ps) for chunk in chunks]
            results = [future.result() for future in concurrent.futures.as_completed(futures)]
        
        # Combine results
        transcript = ' '.join(results)
        return transcript
    
    except Exception as e:
        print(f"Error in speech recognition: {str(e)}")
        raise

video_path = "AI & ML in Finance - Lecture 7 - Data Generation.mp4"
audio_path = "extracted_audio.wav"
enhanced_audio_path = "enhanced_audio.wav"

try:
    # Extract audio from video
    extract_audio_from_video(video_path, audio_path)
    
    # Enhance audio
    enhance_audio(audio_path, enhanced_audio_path)
    
    # Convert audio to text
    text = audio_to_text(enhanced_audio_path)
    
    print("Extracted Text:")
    print(text)
except Exception as e:
    print(f"An error occurred: {str(e)}")
    print("Please make sure you have the required libraries and dependencies:")
    print("pip install pocketsphinx pydub")
    print("Also ensure you have FFmpeg installed on your system.")
finally:
    # Clean up temporary files
    for path in [audio_path, enhanced_audio_path]:
        if os.path.exists(path):
            os.remove(path)
```
